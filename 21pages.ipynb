```python
from bs4 import BeautifulSoup
import pandas
import requests

# 21 pages
l = []
a = []
b = []
cl = []
e = []
f = []
for page in range(0,22):
    base_url = "http://www.escolhaacertada.pt/pt-PT/Imoveis/-/-/-/-/-/-/-/-/-/-/-/-/0/0/20/1/"
    if page == 0:
        continue
    #rint(base_url + str(page))
    r = requests.get(base_url+str(page), headers={
        'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'
    })
    c = r.content
    soup = BeautifulSoup(c, "html.parser")
    all = soup.find_all("div", {"class":"row"})
    #print(len(all))
    for item in all:
        d = {}
        price = soup.find_all("div", {"class":"price"})
        name = soup.find_all("div", {"class":"info"})
        #print(name)
        i_list = soup.find_all("i", class_="icon-area")
        beds = soup.find_all("i", class_="icon-bedrooms")
        bath = soup.find_all("i", class_="icon-bathrooms")
    
    for prices in price:
        prices = prices.text
        prices = " ".join(prices.split())
        a.append(prices)
        d["Preço"] = a
        
    
    for names in name:
        names = names.find_all("a", {"class":"trackevent"})
        names = names[0].text
        b.append(names)
        d["Imóvel"] = b

    for area in i_list:
        area = area.parent.text
        cl.append(area)
        d["Área"] = cl
        
    for bed in beds:
        bed = bed.parent.text
        e.append(bed)
        d["Quartos"] = e
    
    for baths in bath:
        baths = baths.parent.text
        f.append(baths)
        d["Casas de banhos"] = f

    l.append(d)
```

```python
import pandas
for i in l:
    #print(i)
    df = pandas.DataFrame.from_dict(i, orient='index')
    df = df.transpose()
    #df = df[['Imóvel', 'Preço', 'Quartos', 'Área', 'Casas de banhos']]
```

```python
df = df[['Imóvel', 'Preço', 'Quartos', 'Área', 'Casas de banhos']]
```
