```python
import requests
from bs4 import BeautifulSoup

r = requests.get("http://www.escolhaacertada.pt/pt-PT/Imoveis/-/-/-/-/-/-/-/-/-/-/-/-/0/0/20/1/1",
                headers={
                    'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'
                })
c = r.content

soup = BeautifulSoup(c, "html.parser")
all = soup.find_all("div", {"class":"row"})

l = []
a = []
b = []
c = []
e = []
f = []

for i in all:
    d = {}
    price = soup.find_all("div", {"class":"price"})
    name = soup.find_all("div", {"class":"info"})
    i_list = soup.find_all("i", class_="icon-area")
    beds = soup.find_all("i", class_="icon-bedrooms")
    bath = soup.find_all("i", class_="icon-bathrooms")
    
for prices in price:
    prices = prices.text
    prices = " ".join(prices.split())
    a.append(prices)
    d["Preço"] = a
    
for names in name:
    names = names.find_all("a", {"class":"trackevent"})
    names = names[0].text
    b.append(names)
    d["Imóvel"] = b
    
for area in i_list:
    area = area.parent.text
    c.append(area)
    d["Área"] = c
    
for bed in beds:
    bed = bed.parent.text
    e.append(bed)
    d["Quartos"] = e
    
for baths in bath:
    baths = baths.parent.text
    f.append(baths)
    d["Casas de banhos"] = f
    
l.append(d)
```

```python
import pandas
for i in l:
    df = pandas.DataFrame.from_dict(i, orient='index')
```
